```python
# format txt output of ~/development/nlp/nlpie/data/trauma_data/convert.sh 

import re, os, glob, path

directory_to_parse = '/Users/gms/development/nlp/nlpie/data/trauma_data/incoming'

os.chdir(directory_to_parse)

for fname in glob.glob("/Users/gms/development/nlp/nlpie/data/trauma_data/incoming/*.txt"):
    
    # get filename and use for processed output filename
    t = os.path.basename(fname)
    u = t.split('.')[0] + '-v1.txt'
    #print(u)
    
    with open(fname) as f:
        with open('/Users/gms/development/nlp/nlpie/data/trauma_data/out_scrubbed/' + u, 'w') as f2:
            print(f.name)
            
            for line in f:
                a = line.strip() # strip leading and trailing white space
                # b = re.sub('  +','|', a) # strip inner white space >= 2 space replace with '|'
                b = re.sub('  +',' | ', a) # strip inner white space >= 2 space replace with '|'; replaced condition on 1-oct-2018
                
                # clean up "tabular" column heade
                #d = re.sub('Response to', '\n\n', b)
                e = re.sub('Administered\|Medication\|Medication\|Response', '*Administered Prior|Medication Given|Dosage|Units|Route|Response', b)
                f = re.sub('Units EMS Care\|Given\|Dosage\|Units\|Route\|Medication', '\n', e)
                g = re.sub('Prior to this\|Medication\|Medication\|Dosage\|Administered\|to', '\n', f)
                h = re.sub('Location\|Description\|Details', '*Location|Description|Details', g)
                i = re.sub('Primary / Secondary\|Complaint', '*Primary / Secondary|Complaint', h)
                j = re.sub('Abdomen Exam\|Extremities Exam\|Back and Spine Exam', '*Abdomen Exam|Extremities Exam|Back and Spine Exam', i)
                
                # cases for procedure:
                k = re.sub('PTA\?\|Procedure\|Size Attempts\|Response to Procedure', '*PTA?|Procedure|Size Attempts|Response', j)
                l = re.sub('PTA\?\|Procedure\|Size\|Attempts\|Procedure\|Complication', '*PTA?|Procedure|Attempts|Size|Response|Complication', k)
                m = re.sub('PTA\?\|Procedure\|Attempts\|Procedure', '*PTA?|Procedure|Attempts|Response', l)
                n = re.sub('PTA\?\|Procedure\|Attempts\|Procedure\|Complication', '*PTA?|Procedure|Attempts|Response|Complication', m)
                o = re.sub('PTA\?\|Procedure\|Size\|Attempts\|Procedure', '*PTA?|Procedure|Size|Attempts|Response', n)
                p = re.sub('PTA\?\|Procedure\|Size\|Attempts', '*PTA?|Procedure|Size|Attempts', o)
                q = re.sub('PTA\?\|Procedure', '*PTA?|Procedure', p)
               
                # there may be a date of MVC, so scrub it!
                regex = r"\d{2}/\d{2}/\d{4} \d{2}:\d{2}:\d{2}"
                
                r = re.sub(regex, 'dd/dd/dddd dd:dd:dd', q)
               
                #print(i)
                f2.write(r.strip() + '\n')  # python will convert \n to os.linesep

    f2.close()        
```
